# Train-Test Split

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

```python
def simple_linear_regression(input_feature,output):
    sum_x = np.sum(input_feature)
    sum_y = np.sum(output)
    n = len(output)
    slope = ((np.dot(output.reshape(1,-1),input_feature)) - (((sum_x * sum_y)/n)))/((np.sum(np.square(input_feature))) - (sum_x * sum_x)/n)
    intercept = (np.mean(output)) - slope * (np.mean(input_feature))
    return(slope,intercept)
```

```python
input_feature = np.array(train_data['sqft_living']).reshape(-1,1)
output = np.array(train_data['price']).reshape(-1,1)
print(input_feature.shape,output.shape)
```

```python
sqft_slope,sqft_intercept = simple_linear_regression(input_feature,output)
```

```python
def prediction(input_feature,slope,intercept):
    pred_prices = sqft_slope * input_feature + intercept
    return pred_prices

input_feature = train_data['sqft_living']
y_pred = prediction(input_feature,sqft_slope,sqft_intercept)
```

```python
y_pred.reshape(-1,1)
```

```python
print(np.round(prediction(2650,sqft_slope,sqft_intercept),2))
```

```python
def rss(input_feature,output,intercept,slope):
    y_pred = prediction(input_feature,slope,intercept)
    return np.sum(np.square((output - y_pred)))
```

```python
input_feature = np.array(train_data['sqft_living']).reshape(-1,1)
output = np.array(train_data['price']).reshape(-1,1)
print(rss(input_feature,output,sqft_intercept,sqft_slope))
```
