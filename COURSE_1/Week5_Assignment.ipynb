```python
import pandas as pd
import numpy as np
import turicreate
```

```python
songs = pd.read_csv('song_data.csv')
songs
```

```python
songs_sf = turicreate.SFrame(songs)
songs_sf.head()
```

```python
kw = songs_sf[songs_sf['artist'] == 'Kanye West']
ff = songs_sf[songs_sf['artist'] == 'Foo Fighters']
ts = songs_sf[songs_sf['artist'] == 'Taylor Swift']
lg = songs_sf[songs_sf['artist'] == 'Lady GaGa']
```

```python
my_dict = {}
my_dict['Kanye West'] = len(kw['user_id'].unique())
my_dict['Foo Fighter'] = len(ff['user_id'].unique())
my_dict['Taylor Swift'] = len(ts['user_id'].unique())
my_dict['Lady GaGa'] = len(lg['user_id'].unique())
my_dict
```

```python
popular_artists = songs.groupby('artist')['listen_count'].sum()
```

```python
print(np.argmax(popular_artists))
print(max(popular_artists))
```

```python
print(np.argmin(popular_artists))
print(min(popular_artists))
```
