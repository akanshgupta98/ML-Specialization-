```python
import turicreate
import pandas as pd
```

# Reading the dataset

```python
sf = turicreate.SFrame('amazon_baby.sframe')
sf.head()
```

# Exploring The Data

```python
sf.show()
```

```python
sf.groupby('name',operations={'count':turicreate.aggregate.COUNT()}).sort('count',ascending=False)
```

# Count Vector

```python
sf['word_count'] = turicreate.text_analytics.count_words(sf['review'])
sf.head()
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
def awesome_count(x):
    count = 0
    if('awesome' in x):
        count = x['awesome'] 
    return count
        
```

```python
%%time
res = sf['word_count'].apply(awesome_count)
```

```python
sf['awesome_wc'] = sf['word_count'].apply(awesome_count)
```

```python
sf.head()
```

```python
def selected_wc(x):
    my_dict = {}
    for key in x.keys():
        if(key in selected_words):
            my_dict[key + '_wc'] = x[key] 
    return my_dict
```

```python
%%time
res = sf['word_count'].apply(selected_wc)
```

```python
df = pd.DataFrame(res)
df.fillna(0,inplace=True)
```

```python
df
```

```python
new_sf = turicreate.SFrame(df)
```

```python
new_sf
```

```python
import numpy as np
np.sum(df)
```

## WOW -> Least Used. Great -> Max Used

```python
sf
```

```python
sf.remove_column('awesome_wc',inplace=True)
sf
```

```python
sf = sf.add_columns(new_sf)
```

```python
sf.shape
```

# Removing 3.0 ratings and adding sentiment column

```python
sf = sf[sf['rating'] != 3.0]
```

```python
sf.shape
```

```python
sf['sentiment'] = sf['rating'] >=4.0
```

```python
sf.shape
```

```python
sf['sentiment']
```

# Train-Test Split

```python
train_data,test_data = sf.random_split(seed=0,fraction=0.8)
print(train_data.shape,test_data.shape)
```

```python
features = selected_words
features
```

```python
def fix_col_names(x):
    return x + '_wc'
features = list(map(fix_col_names,features))
```

```python
features
```

```python
train_data['sentiment']
```

```python
train_data[features].shape
```

# Training With Full Words

```python
sentiment_model = turicreate.logistic_classifier.create(train_data,target='sentiment',validation_set=test_data,features=['word_count'])
```

# Training with Just Some Words

```python
selected_words_model = turicreate.logistic_classifier.create(train_data,target='sentiment',validation_set=test_data,features=features)
```

```python
selected_words_model.coefficients.sort('value')
```

* Most Positive Weight = Awesome. 
* Most Negative Weight = Horrible

```python
selected_words_model.evaluate(test_data)
```
